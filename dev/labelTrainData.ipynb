```python
import sys
sys.path.append('../')

import lib.utils as utils
import os

import cv2
import numpy as np
```

```python
def convertImg(image, size=(32,32)):
    image = cv2.resize(image, size, interpolation=cv2.INTER_AREA)
    converted = np.zeros(shape=(size[0], size[1], 1))
    for y in range(0, len(image)):
        for x in range(0, len(image[y])):
            converted[y][x][0] = image[y][x]

    converted = converted/255
    return converted
```

```python
data = os.listdir('./../resources/trainData')
print(data)

class_names = os.listdir('./../resources/trainData_new')
print(class_names)
```

```python
counter = [0 for i in range(0, len(class_names))]

for i in range(0, len(class_names)):
    images = os.listdir('./../resources/trainData_new/'+class_names[i])
    counter[i] = len(images)

print(counter)

for i in range(0, len(data)):
    image = utils.loadImage(path='./../resources/trainData/', name=data[i], resize=(32,32))

    large_image = cv2.resize(image, (256,256))
    cv2.imshow('name', large_image)


    path = './../resources/trainData_new/'
    name = ''

    #print(counter)
    k = cv2.waitKey(0)
    #print(k)


    if k == 99:
        path += 'capacitor'
        ind = class_names.index('capacitor')
        name += str(counter[ind])
        counter[ind] += 1
    if k == 114:
        path += 'resistor'
        ind = class_names.index('resistor')
        name += str(counter[ind])
        counter[ind] += 1
    if k == 103:
        path += 'ground'
        ind = class_names.index('ground')
        name += str(counter[ind])
        counter[ind] += 1
    if k == 105:
        path += 'inductor'
        ind = class_names.index('inductor')
        name += str(counter[ind])
        counter[ind] += 1
    if k == 118:
        path += 'voltage'
        ind = class_names.index('voltage')
        name += str(counter[ind])
        counter[ind] += 1
    name += '.png'

    # closing all open windows
    cv2.destroyAllWindows()

    path += '/'

    #print(path)

    utils.saveImage(name, image, path=path, transpose=False, invert=False, color=False)
```
